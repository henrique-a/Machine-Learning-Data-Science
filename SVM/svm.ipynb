```python
import numpy as np
from sklearn import model_selection, preprocessing
import pandas as pd
```

```python
df = pd.read_csv('../Data/breast_cancer.csv')
```

```python
df.replace('?', -99999, inplace=True)
df.drop(['id'], 1, inplace=True)
```

```python
X = np.array(df.drop(['class'], 1))
X = X.astype(np.float)
X = preprocessing.scale(X)
y = np.array(df['class'])
for i in range(y.size):
    if y[i] == 2:
        y[i] = -1
    elif y[i] == 4:
        y[i] = 1
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2)
```

```python
# Perform a stochastic gradient descent
def svm_sgd(X, y):
    w = np.zeros(X[0].size) #Initialize our SVMs weight vector with zeros (3 values)
    eta = 1 #The learning rate
    epochs = 1000 #how many iterations to train for

    #training
    for epoch in range(1,epochs):
        errors = 0
        for i in range(len(X)):
            if (y[i] * np.dot(X[i], w)) < 1: #misclassification
                w = w + eta * ((X[i] * y[i]) + (-2  *(1/epoch)* w)) #misclassified update for ours weights
                errors += 1
            else:
                w = w + eta * (-2  *(1/epoch)* w) #correct classification, update our weights

        accuracy = (1 - errors/len(X)) * 100
        print('{0:.2f}%'.format(accuracy))
    
    return w
```

```python
w = svm_sgd(X_train, y_train)
```

```python
# Evaluate with test set
errors = 0
for i in range(len(X_test)):
    if (y_test[i] * np.dot(X_test[i], w)) < 1:  #misclassification

        errors += 1
accuracy = (1 - errors/len(X_test)) * 100
print('{0:.2f}%'.format(accuracy))
```
