```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[1,1,1,1,1,1], [1,2,3,4,5,6]]).reshape(2,6)
y = np.array([5,4,6,5,6,7]).reshape(1,6)
N,M = x.shape # Number of samples and number of features
```

```python
plt.plot(x[1].flatten(), y.flatten(), 'bo')
plt.xlabel("$x$", fontsize = 18)
plt.ylabel("$y$", fontsize = 18)
plt.grid()
plt.show()
```

```python
# Placeholder
tf.reset_default_graph()
features = tf.placeholder(tf.float32, shape=(N,M)) # placeholder for features
target = tf.placeholder(tf.float32, shape=(1,M)) # placeholder for targets
```

```python
# Make predictions
weights = tf.get_variable("w", shape=(1,N), dtype=tf.float32) # variable for weights
predictions = weights @ features
```

```python
# Define loss function
loss = tf.reduce_mean((target - predictions)**2)
```

```python
# Optimizer
optimizer = tf.train.GradientDescentOptimizer(0.01)
step = optimizer.minimize(loss)
```

```python
# Solve Linear Regression
s = tf.InteractiveSession()
s.run(tf.global_variables_initializer())


for i in range(1000):
    _, loss_new, weights_new = s.run([step,loss,weights],feed_dict={features:x, target:y})
    print(loss_new)

s.close()
```

```python
pred = weights_new @ x
plt.plot(x[1].flatten(), y.flatten(), 'bo')
plt.plot(x[1].flatten(), pred.flatten(), 'r-')
plt.xlabel("$x$", fontsize = 18)
plt.ylabel("$y$", fontsize = 18)
plt.grid()
plt.show()
```
